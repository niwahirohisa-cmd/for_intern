```python
import pandas as pd;import numpy as np;
import glob;from datetime import datetime;from datetime import timedelta
from io import StringIO

#サンプルデータを作成
data = """
date,temp
2022/08/01 00:00:00,25
2022/08/01 00:00:00,28
2022/08/01 00:00:00,30
2022/08/02 00:00:00,26
2022/08/02 00:00:00,32
2022/08/02 00:00:00,38
2021/08/01 00:00:00,27
2021/08/01 00:00:00,32
2021/08/01 00:00:00,34
2021/08/02 00:00:00,29
2021/08/02 00:00:00,35
2021/08/02 00:00:00,40
"""

df = pd.read_csv(StringIO(data))

# データフレームを表示
print(df)

#どこかに保存してあるcsvファイルを読み込んでも良いです
#df = pd.read_csv("/usr/amoeba/pub/for_intern/data/test.csv")
```

```python
#時間型に変更
df['date']=pd.to_datetime(df['date'])
```

```python
#年を抽出してカラムに追加
df["year"]=df["date"].dt.year
#月を抽出してカラムに追加
df["month"]=df["date"].dt.month
#日付を抽出してカラムに追加
df["day"]=df["date"].dt.day
df
```

```python
#年別、月別、日別の最高気温を算出。max()部分は,min(),mean(),sum()なども可
df2 = df.groupby(["year","month","day"],as_index=False)["temp"].max()
df2
```

```python
#年別、月別の最高気温の平均値を計算
df2.groupby(["year","month"],as_index=False)["temp"].mean()
```
