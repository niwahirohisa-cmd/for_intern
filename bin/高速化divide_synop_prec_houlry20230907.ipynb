```python
#高速化バージョン
import pandas as pd
import numpy as np
from datetime import timedelta
import re

# パス文字列
#input_path = "/usr/amoeba/pub/for_intern/data/1001_processed_data.csv"
# 正規表現で数値部分を抽出
#match = re.search(r"(\d+)_processed_data.csv", input_path)  # 'path' を 'input_path' に変更
#synopid = match.group(1)
synopid = 47662

input_path = "/usr/amoeba/pub/for_intern/bin/sample_data/input_divide_synop_houlry_prec.csv"


df_org = pd.read_csv(input_path)
df_org = df_org[df_org['PrecipitationTerm'] != 0]

# NumPy arrays
df_org["localtime"] = pd.to_datetime(df_org["localtime"])
localtimes = df_org["localtime"].values.astype('datetime64[h]')
precipitation_terms = df_org["PrecipitationTerm"].values
precipitations = df_org["Precipitation"].values

unique_times = np.unique(localtimes)

all_utcs = []
all_precipitations = []

for utc in unique_times:
    mask = localtimes == utc
    hour_interval = precipitation_terms[mask][0]
    total_prec = precipitations[mask][0]
    prec_per_hour = total_prec / hour_interval
    
    valid_utcs = np.array([utc - np.timedelta64(i, 'h') for i in range(hour_interval)])
    all_utcs.extend(valid_utcs)
    all_precipitations.extend([prec_per_hour] * hour_interval)

# Convert back to DataFrame for sorting and saving
df_all = pd.DataFrame({
    'localtime': all_utcs,
    'Precipitation': all_precipitations
}).sort_values(by="localtime", ascending=True)

# 作成したファイルをローカルに保存
#output_path = '/usr/amoeba/pub/for_intern/data/output2{}.csv'.format(synopid)
output_path = '/usr/amoeba/pub/for_intern/bin/sample_data/output_divide_synop_houlry_prec.csv'
df_all.to_csv(output_path, header=True, index=False)
```
